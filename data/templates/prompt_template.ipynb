# Prompt & Evaluation Templates

## Objective
> This notebook documents the exact prompt templates used for belief elicitation, adversarial pressure, debate exchanges, and judge evaluation.

These templates are provided to:
* Ensure experimental reproducibility
* Clarify the structure of the sandbox debate protocol
* Allow future extensions or controlled variations

## High-Level Debate Structure

Each experiment follows the same sequence:

1. Initial Belief Elicitation
2. Pressure Prompt (Argument A)
3. Challenger Rebuttal (Argument B)
4. Proposer Defense (Argument C)
5. Challenger Rebuttal (Argument D)
6. Final Proposer Position (Argument E)
7. Independent Judge Evaluation

Roles are explicitly assigned and may be swapped across runs.
